```python
dict1={}
a=[]
#getting all the mean values from the specific province area into a dictionary
for i in df['province'].unique():
  df2=df[df.province==i]
  for j in ['roads','proads','cres','csor','organic','paper','glass','wood','metal','plastic','raee','texile','other','s_landfill','s_wteregio','geo','wage','finance','urb','wden','pden','sea','isle','alt','area']:
    a=df2[j].mean()
    dict1.update({i+'-'+j:a})
#replacing the dictionary values to the column null values
print(dict1)
c=0
#df1['cres'][df1['cres'].isnull()]
for i in range(0,df.shape[0]):
  for j in ['roads','proads','cres','csor','organic','paper','glass','wood','metal','plastic','raee','texile','other','s_landfill','s_wteregio','geo','wage','finance','urb','wden','pden','sea','isle','alt','area']:
    if(np.isnan(df.loc[i,j])):
      c=c+1
      df.loc[i,j]=dict1[df.iloc[i,1]+ '-' + j]
```

```python
#replacing the null values which are still there with mean value of the column
for j in ['wood','roads','proads','finance','gdp','organic']:
    print(df[j].mean())
    df[j]=df[j].fillna(df[j].mean())
df.isnull().sum()
```

```python
#replacing the string null values with sensible name

a=df['name'][df.isna().any(axis=1)].index
a=a.tolist() 
k=0
for i in a:
    k=k+1
    df.loc[a,'name']='unknown'+str(k)
```
